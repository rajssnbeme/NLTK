Filtering Stop Words


**Stop words** are a set of commonly used words in a language. **Examples** of stop words in **English are “a,” “the,” “is,” “are,” etc.**

The following is a broad list of stopword categories:
 - Common Stopwords
          Examples include “the,” “is,” “in,” “for,” “where,” “when,” “to,” “at,” etc.
  - Custom Stopwords
          For example, in a medical context, words like “patient” or “treatment” might be considered as custom stopwords.
  - Numerical Stopwords
  -Single-Character Stopwords
  -Contextual Stopwords




```python
nltk.download("stopwords")
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
print(stopwords.words('english'))
```

```python
worf_quote = "Sir, I protest. I am not a merry man!"
```

```python
words_in_quote = word_tokenize(worf_quote)
words_in_quote
```

```python
stop_words = set(stopwords.words("english"))
stop_words
```

```python
filtered_list = []
```

```python
for word in words_in_quote:
  if word.casefold() not in stop_words:
             filtered_list.append(word)
```

```python
filtered_list
```

```python
for word in words_in_quote:
  if word not in stop_words:
         filtered_list.append(word)
```

```python
filtered_list
```

Alternatively, you could use a list comprehension to make a list of all the words in your text that aren’t stop words:

```python
filtered_list = [
...     word for word in words_in_quote if word.casefold() not in stop_words
... ]
filtered_list
```

Stemming

       The process of removing affixes from a word so that we are left with the stem of that word is called stemming.
             For example, the words “helping” and “helper” share the root “help.”![index1.jpg](![Capture.PNG](![Capture.PNG](Capture.PNG)))

       

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```python
stemmer = PorterStemmer()
```

```python
string_for_stemming = """
... The crew of the USS Discovery discovered many discoveries.
... Discovering is what explorers do."""
```

```python
words = word_tokenize(string_for_stemming)
words
```

```python
stemmed_words = [stemmer.stem(word) for word in words]
stemmed_words
```
